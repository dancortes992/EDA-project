### `order_products` data frame

Use `min` and `max`functions on `'add_to_cart_order'` column.

```python
# What are the min and max values in this column?
min_value = df_order_products['add_to_cart_order'].min()
max_value = df_order_products['add_to_cart_order'].max()

print(f"Minimum value: {min_value}, Maximum value: {max_value}")
```

We filter `add_to_cart_order` where values are `NaN` and save all unique `'order_id'`s where in occurs in an array. 

```python
# Save all order IDs with at least one missing value in 'add_to_cart_order'
missing_order_ids = df_order_products[df_order_products['add_to_cart_order'].isna()]['order_id'].unique()
print(missing_order_ids)
```

 We can group by `'order_id'` and check if said ID's appear in our previous array. We use `size` to count the number of elements in these ID's. 

```python
# Do all orders with missing values have more than 64 products?
df_order_size_ids = df_order_products[df_order_products['order_id'].isin(missing_order_ids)].groupby('order_id').size()
def greater_64(series):
    for value in series:
        if value > 64:
            return True

print(df_order_size_ids)
print()
print("More tan 64 products in every order:", greater_64(df_order_size_ids))
```

To check if every item in the filtered array has more then 64 products, we can create a function `greater_64` that checks if this statment is True.

We fill these missing values with 999 to aviod errors in analysis and convert the `'add_to_cart_order'` to integer using `astype`.

```python
# Replace missing values with 999 and convert column to integer type
df_order_products['add_to_cart_order'] = df_order_products['add_to_cart_order'].fillna(999)
missing_add_to_cart_values = df_order_products[df_order_products['add_to_cart_order'] == 999]
df_order_products['add_to_cart_order'] = df_order_products['add_to_cart_order'].astype('int')

print(missing_add_to_cart_values)
df_order_products.info(show_counts=True)
```

We can plot the time of the day with most orders by obtaining the `value_counts` of `'order_hour_of_day'`.

### What are the top 20 items that are reordered most frequently?

To obtain the number of reorders products we must filter order products dataset on the basis of `'reordered'` == 1 (product reordered). And call `value_counts` to obtain the count of reordered times.

```python
reorders = df_order_products.query("reordered == 1")[['product_id', 'reordered']]
reorder_count = reorders['product_id'].value_counts().head(20).reset_index()
reorder_count.columns = ['product_id', 'reorder_count']
print(reorder_count)
```

To display product names we merge our reorder count with products dataset on `'product_id'`.

```python
top_20_reorders = reorder_count.merge(df_products[['product_id', 'product_name']], on='product_id', how='left')
print(top_20_reorders)
```

We can see the majority of the top 20 reordered products are also most of the top 20 products bought in general. 

### For each product, what proportion of its orders are reorders?

We must obtain a count of `'order_id'` for each `'product_id'` and divide by the sum of `'reordered'` items for each `'product_id'`. This will result in the proportion of reorders per product orders.

```python
total_orders = df_order_products.groupby('product_id')['order_id'].count()
total_reorders = df_order_products.groupby('product_id')['reordered'].sum()
product_reorder_proportion = (total_reorders / total_orders) * 100
product_reorder_proportion_merged = product_reorder_proportion.to_frame().merge(df_products[['product_id', 'product_name']], on='product_id', how='left')
product_reorder_proportion_merged.columns = ['product_id', 'percent', 'product_name']

print("Reorder proportion/percent:")
print()
print(product_reorder_proportion_merged)
```

### For each customer, what proportion of their products ordered are reorders?

By merging order_products dataset with the orders dataset on `'order_id'` we obtain a dataset that contains both item reorders and user IDs. This allows us to obatin the `mean` of reordered items by grouping per costumer (`'user_id'`).

### What are the top 20 items that people put in their carts first?
